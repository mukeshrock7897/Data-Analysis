# **8. Merging and Joining DataFrames**

**Merging DataFrames**
* Merge two DataFrames on a key.

```python
df1 = pd.DataFrame({'key': ['A', 'B', 'C'], 'value1': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['A', 'B', 'D'], 'value2': [4, 5, 6]})
merged_df = pd.merge(df1, df2, on='key')
print(merged_df)
```

**Joining DataFrames**
* Join two DataFrames on their index.

```python
df1 = pd.DataFrame({'value1': [1, 2, 3]}, index=['A', 'B', 'C'])
df2 = pd.DataFrame({'value2': [4, 5, 6]}, index=['A', 'B', 'D'])
joined_df = df1.join(df2, how='inner')
print(joined_df)
```

# **10. Advanced Data Analysis**

**Creating a pivot table**
* Create a pivot table for summary data.

```python
# Check the existing columns in your DataFrame
print(df.columns)

# If 'City' is not present, create it with sample data
df['City'] = ['New York', 'London', 'Tokyo', 'Paris']  # Replace with your actual city data

# Now try the pivot operation again
df.pivot_table(values='Age', index='Name', columns='City', aggfunc='mean')
print(df)
```

**Resampling time-series data**
* Resample time-series data.

```python
# Check if 'Date' is in the original DataFrame
if 'Date' in df.columns:
    # Convert 'Date' to datetime if it's not already
    df['Date'] = pd.to_datetime(df['Date'])

    # Select numeric columns including 'Date'
    df_numeric = df.select_dtypes(include=['number', 'datetime'])

    # Set 'Date' as index and resample by month, then calculate the mean of numeric columns
    result = df_numeric.set_index('Date').resample('M').mean()
    print(result)
else:
    print("The 'Date' column is not present in the DataFrame.")
```
